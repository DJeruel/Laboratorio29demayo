# Carga y lectura de datos csv

```python
#importar librería
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
```

```python
path ="C:/Users/JERUEL TOMA/Downloads"       #direccion donde se guardan los archivos
file = "datosincompletos2020perfiles.csv"       #nombre del archivo a cargar
pathdata = os.path.join(path,file)   #direccion de los datos donde se cargan
```

```python
#Lectura csv
data = pd.read_csv(pathdata)
#Impresion CSV
#data.head ()
#Impresion completa CSV
data
```


# 1) crear un archivo CSV completo fusionando el archivo original con el nuevo creado.

```python
# Numero de datos en bd
ndatos = 25

# a) Edad (su  rango debe ir entre los 20 a 40 años  (con desviación están de  4 años).
edad = np.random.uniform(20,40,ndatos) + 4*np.random.randn(ndatos)

#b) Años de experiencia como mínimo 2 como máximo 20.
aniosdeexp = np.random.uniform(2,20,ndatos)

#c) Años de estudio (tome en cuenta que como mínimo tienen técnico y máximo maestría ).
aniosdeestudio = np.random.uniform(2,5,ndatos)


nuevatabla = pd.DataFrame(
    {
        'edad':edad,
        'años de experiencia':aniosdeexp,
        'Años de estudio':aniosdeestudio
    }
)

nuevatabla

```

# Combinacion de tablas (dadas por catedratico + creadas por estudiantes)

```python
combinaciontablas = pd.concat([data, nuevatabla], axis=1)
combinaciontablas
```

# 2) Realice un histograma para mostrar el salario máximo y  el salario mínimo.

```python
salariosmaxymin = []

listaingresos = combinaciontablas["ingresos"].values.tolist()
Salmax = max(listaingresos)
Salmin = min(listaingresos)
salariosmaxymin.append(max(listaingresos))
salariosmaxymin.append(min(listaingresos))

# espacio de impresion
%matplotlib inline
plt.hist(salariosmaxymin)


```

# 4) Creer 2 diagramas de dispersión

### para verificar sí la edad tiene influencia con los ingresos

```python
sns.set_theme(color_codes=True)
sns.regplot(x="edad", y="ingresos", data=combinaciontablas);
```

###### Interpretación: Las personas que tienen de 25 a 35 años son las que tiene mejor ingreso ( de 10,000 a 12,000) , pero cuando las personas que cuentan con edad  superior a los 40 años su sueldo disminuye. entonces en efecto, la edad si tiene influencia respecto al sueldo.

### sí los años de experiencia tienen que ver con los ingresos

```python
sns.set_theme(color_codes=True)
sns.regplot(x="años de experiencia", y="ingresos", data=combinaciontablas);
```

###### Interpretación: En efecto, las personas con mayor experiencia son las personas con mayor sueldo, por ejemplo, observamos que las personas que tienen de 12 a 18 años de experiencia tiene un sueldo de 4 a 12 mil, mientras que las personas que cuentan con 4 a 9 años de experiencia tienen un sueldo de 4 a 8 mil.
